```python
import yaml
import pandas as pd
from copy import deepcopy
from src.train import train_model

with open("config.yaml") as f:
    base_config = yaml.safe_load(f)

results = []
```

```python
# 1️⃣ Baseline
config1 = deepcopy(base_config)
train_loss, val_loss, val_acc = train_model(config1, "baseline")
results.append(["Baseline", "Default BERT", train_loss, val_loss, val_acc])
```

```python
# 2️⃣ Increase Dropout
config2 = deepcopy(base_config)
config2["dropout_rate"] = 0.3
train_loss, val_loss, val_acc = train_model(config2, "dropout_0.3")
results.append(["Exp2", "Dropout 0.3", train_loss, val_loss, val_acc])
```

```python
# 3️⃣ Freeze BERT
config3 = deepcopy(base_config)
config3["freeze_bert"] = True
train_loss, val_loss, val_acc = train_model(config3, "freeze_bert")
results.append(["Exp3", "Freeze BERT", train_loss, val_loss, val_acc])
```

```python
# 4️⃣ Reduced Learning Rate
config4 = deepcopy(base_config)
config4["learning_rate"] = 5e-5
train_loss, val_loss, val_acc = train_model(config4, "reduced_lr")
results.append(["Exp4", "LR 5e-5", train_loss, val_loss, val_acc])
```

```python
# 5️⃣ More Epochs
config5 = deepcopy(base_config)
config5["epochs"] = 3
train_loss, val_loss, val_acc = train_model(config5, "more_epochs")
results.append(["Exp5", "Epochs 3", train_loss, val_loss, val_acc])
```

```python
import pandas as pd

results = [
    ["Baseline", "Default BERT", 1.6420, 1.6444, 0.2431],
    ["Exp2", "Dropout 0.3", 1.7148, 1.6113, 0.2265],
    ["Exp3", "Freeze BERT", 1.6291, 1.5957, 0.2451],
    ["Exp4", "LR 5e-5", 1.6276, 1.5799, 0.2775],
    ["Exp5", "Increase Epochs", 1.6325, 1.5958, 0.2377]
]

df = pd.DataFrame(results, columns=[
    "Experiment",
    "Modification",
    "Train Loss",
    "Validation Loss",
    "Validation Accuracy"
])

df.to_csv("results/results.csv", index=False)

df
```
